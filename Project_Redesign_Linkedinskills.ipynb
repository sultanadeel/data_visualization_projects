The crucial takeaway of the dataset is that Data skills are becoming more popular and increasing in demand across all the major countries used in the dataset, and from 2014 to 2016, the data skills have moved up in terms of their respective ranks

Our utilmate Claim using this Data Wrangling and subsequent Tableau Visualization is that "Statistical Analysis, Data Mining and Cloud Computing are becoming top ranked LinkedIn skills and therefore their demand is increasing."

# Plotting Box Plot to analyze the relationship & data distribution between Rank and Year entities

```python
plt.rcParams["figure.figsize"] = (8,8)
df1.boxplot(column='Rank', by = 'Year')
```

The above boxplot shows that in the years 2014 and 2015, the distribution of skills was uniform as all the ranks from 1 to 25
were evenly distributed as shown by the respective medians, whiskers, and interquartile ranges of the boxplot. However, in the
year 2016 , the median rank is around 6 and the upper quartile range of the boxplot is Rank 10. The important analysis from the
year 2016 is that, all the top raked skills were more popular in the year 2016 and most of the below ranked skills were either
not in demand or been entirely out of the scope of the analysis for the LinkedIn Top skills. However, 2016 is the year of changing
trend and pattern, as the Data and Technology skills dominated the mainstream skillset for the different countries being analyzed

# Analyzing Country counts using bar charts

```python
df1.Country.value_counts().plot(kind='bar')
plt.figure(figsize=(2,2))
```

```python
from collections import Counter
plt.figure(figsize=(5,5))
c = pd.Series(Counter(df1.Country.str.split(',').sum()))
c.plot(kind='barh',title='Country Count',color="blue")
```

The country count bar chart shows that most of the countries have appeared over 60 times in the dataset, indicating the fact that
they are very popular on LinkedIn search in representing the respective skill. However, some countries like Singapore, Ireland,
and China have occurred only 35 times in the dataset

```python
from collections import Counter
plt.figure(figsize=(22,35))
#plt.rcParams["figure.figsize"] = (90,15)
c = pd.Series(Counter(df1.Skill.str.split(',').sum()))
c.plot(kind='barh', title='Skill Count', color=['blue'])

```

The above bar chart shows the counts of skills, and indicates the no.of times a respective skill is mentioned in the LinkedIn
Top Skills Dataset. The advantage of this plot is to quickly analyze the frequency of the skill in consideration. Higher the 
frequency and occurrence of the skill, higher is the popularity and demand of that skill. The biggest skill count is for
Statistical Analysis & Data Mining around 41, which supports our claim that Data Skills are increasing in demand and popularity
